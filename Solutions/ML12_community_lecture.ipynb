```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F","G"])
G.add_edges_from([("A","B"),("A","C"),("B","C"),("B","D"),("D","E"),("D","F")
                 ,("D","G"),("E","F"),("F","G")])
G_pos = {"A": (-2, 0),
       "B": (-1, 0),
       "C": (-1.5, -1),
       "D": (1, 0),
       "E": (3, 0),
       "F": (3,-1 ),
       "G": (1, -1)}
colors=['red', 'blue', 'red','blue','green', 'green', 'green']
# pos = nx.spring_layout(G)
fig, ax = plt.subplots(figsize=(8, 3))
nx.draw_networkx(G, with_labels=True, font_weight='bold',node_color=colors,pos=G_pos,ax=ax)
plt.show()
```

```python
from math import comb
no_edges=nx.number_of_edges(G)
print(no_edges)
cal_density=nx.number_of_edges(G)/comb(nx.number_of_nodes(G),2) 
print(f'calculated Density={cal_density} ')
print(f' Density={nx.density(G)} ') 
print(nx.number_of_nodes(G))
print(comb(nx.number_of_nodes(G),2))
cal_betweenness=12/comb(nx.number_of_nodes(G),2) 
print(f'Calculated betweeness for edge B and D is= {cal_betweenness} ')

nx.edge_betweenness_centrality(G)
print(nx.edge_betweenness_centrality(G))
```

```python
print(nx.conductance(G, ['A','B','C'],['D','E','F','G']))
print(nx.conductance(G, ['D','E','F','G']))
print(nx.conductance(G, ['A','B','C']))
```
